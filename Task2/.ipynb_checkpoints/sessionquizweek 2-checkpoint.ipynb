2.15

```python
night1 = [1, 3, 2, 4, 0, 3, 7, 5, 2, 9, 2, 4, 3]
night2 = [0, 0, 2, 4, 2, 4, 3, 2, 1, 1, 1, 2, 0]
set_night1 = set(night1)
set_night2 = set(night2)
inters = set.intersection(set_night1, set_night2)
print(f"Intersection night1, night2: {inters}")
if 5 in inters:
    print("5 is an element in intersection list")
else:
    print("5 is not an element in intersection list")
if 5 in set_night1:
    print("5 is a member of set list night1 ")
else:
    print("5 is not a member of set list night1")
```
