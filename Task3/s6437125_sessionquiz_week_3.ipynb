```python
# Masses of solar system (in kg)

masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

Mass_of_moon = 7.349e+22


# Create a list for the masses of solar system higher than mass of moon

Filtered_masses = []

for M in masses:
    if M > Mass_of_moon:
        Filtered_masses.append(M)
        
print(f"list of masses higher than mass of the moon: {Filtered_masses}")


# Create a list of the last five masses
indices = slice(6,11,1)
Last_five_masses = masses[indices]
print(f"list of last five masses: {Last_five_masses}")


# Calculate the sum of the last five masses
s = sum(masses[indices])
n = len(masses[indices])
print(f"average of the last five masses: {s/n}")
```

```python
import math 
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

escape_velocities = {}

# Create the masses with two numbers after the dot
for i in range(len(masses)):
    M = masses[i]
    r = radii[i]
    V_esc = math.sqrt(2*G*M/r) 
   
    print(f"{names[i]} mass {M:.2e} kg") 


# Calculate the escape velocities    
    escape_velocities[names[i]] = V_esc
print()
print(f"escape velocities dictionary: {escape_velocities}")




```

```python
import math
from scipy.constants import pi

# defining cylinder

def cylinder(r, h):
    A = 2*pi*r**2 + 2*pi*r*h
    V = pi*r**2*h
    return A, V

# create input

r = float(input(f"input a value for radius: "))
h = float(input(f"input a value for height: "))

# calculate ratio

A, V = cylinder(r, h)
ratio = A/V

print(f"area of cylinder is: {A} m^2")
print(f"volume of cylinder is: {V} m^3")
print(f"ratio of area and volume is: {ratio}")
```

```python
import numpy as np

#generate data
X = np.random.normal(loc=0,scale=5,size=10000)

#function stdev that returns standard deviation of input array
def stdev(arr): 
    return np.sqrt(np.sum((arr-np.sum(arr)/len(arr))**2)/len(arr))


#create a one-liner
stdev_oneliner = lambda arr: np.sqrt(np.mean((arr - np.mean(arr))**2))

#comparing
print(f"stdev: {stdev(X)}")
print(f"numpy.std: {np.std(X)}")
print(f"stdev oneliner: {stdev_oneliner(X)}")
```

```python
import numpy as np

#data of distances and velocities
x = np.array([18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]) #distances
y = np.array([2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]) #velocities

#compute means
X = np.mean(x)
Y = np.mean(y)

#compute b(slope)
b = np.sum((x-X)*(y-Y))/np.sum((x-X)**2)

#compute a(intercept)
a = Y - b*X

#results
print(f"the intercept is: {a} km/s")
print(f"the slope is: {b} km/s/Mpc")

#fitted equation
print(f"fitted linear relation: y = {a} + {b} x")
    
```

```python
import matplotlib.pyplot as plt
import math

#int x values -5 to 6
x_range = range(-5,6)
y_range = [(math.exp(x) + math.exp(-x))/2 for x in x_range]


plt.plot(x_range, y_range, marker = "o", color = "red", label = "y = cosh((e^x+e^-x)/2)")

#adjustments plot
plt.title("Catenary")
plt.xlabel("Values for x", fontsize=11)
plt.ylabel("Values for y", fontsize=11)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid()
plt.show()
```
